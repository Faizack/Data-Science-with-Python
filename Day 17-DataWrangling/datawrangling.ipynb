## Data Normalization
  - Uniform the data
  - they have same impact 
  - aik machli samundar main or aik jar main

```python
ks4=k2[['age','fare']]
ks4.head()

ks5=k2[['age','fare']]
ks6=k2[['age','fare']]
ks7=k2[['age','fare']]
```

- above data is really in wide range and we need to normalize and hard to compare
- Normalization change the values to range of 0 to 1(Now both variable has similar influence on our model)

### **Method of Normalization**

1. Simple Feature Scaling 
    - x(new)=x(old)/x(max))
2. Min-Max Method
3. Z-Score (Standard Score) -3 to + 3
4. Log Transformation
        

```python
#   1. Simple Feature Scaling 
#     - x(new)=x(old)/x(max))
ks4['fare']=ks4['fare']/ks4['fare'].max()
ks4.head()
```

```python
# Min-Max Method
#- x(new)=x(old)-x(min)/(x(max))-(x(min)))

ks5['fare']=ks5['fare']-(ks5['fare'].min())/(ks5['fare'].max())-(ks5['fare'].min())
ks4
```

```python
ks6.head()
```

```python
ks6=sns.load_dataset('titanic')
ks6.head()
```

```python
sns.histplot(data=ks6, x="fare",kde=True)
```

```python
# Z-score (Standard Score)
# x-x(mean)/x(std)

ks6['fare']=(ks6['fare']-ks6['fare'].mean())/ks6['fare'].std()
ks6.head()
```

```python
sns.histplot(data=ks6, x="fare",kde=True)
```

```python
# Log Transformation
ks7['fare']=np.log(ks7['fare'])
ks7.head()
```
